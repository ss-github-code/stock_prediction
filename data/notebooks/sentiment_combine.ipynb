```python
import pandas as pd
import glob
```

```python
path = "/content/drive/MyDrive/sent_data/sentiment_data"
all_files = glob.glob(path + "/*.csv")
```

```python
bucket = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    bucket.append(df)

final_df = pd.concat(bucket, axis=0, ignore_index=True)
final_df.rename(columns={"Unnamed: 0": "date"}, inplace=True)
final_df.sort_values(by=['date'], inplace=True)
final_df.drop_duplicates(inplace=True)
final_df.set_index('date', inplace=True)
final_df = final_df.loc[~(final_df==0).all(axis=1)]
# final_df['date'] = pd.to_datetime(final_df['date'], format="%Y/%m/%d")

# final_df = pd.to_datetime(final_df.date)
```
